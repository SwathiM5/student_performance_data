### Bar Plot

```python
sns.barplot(x='gender', y='overall_score', data=std_per_dt)
plt.title("Gender vs Overall Score")

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#Setting Lines thinner:
for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)

plt.show()
```

```python
plt.Figure(figsize=(8,5))
sns.scatterplot(x='study_hours_per_day', y='overall_score', data=std_per_dt, alpha=0.5)
plt.title("Study Hours vs Overall Score")
plt.xlabel("Study hours per day")
plt.ylabel("overall_score")
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=std_per_dt,
    x="study_hours_per_day",
    y="overall_score",
    hue="gender",
    style="gender",
    palette={"Male": "Green", "Female": "Yellow"},
    alpha=0.5,
    s=40
)

plt.title("Study Hours vs Final Score by Gender")
plt.xlabel("Study Hours per Day")
plt.ylabel("Overall Score")

plt.legend(title="Gender")

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
